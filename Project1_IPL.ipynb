```python
import pandas as pd
import numpy as np
```

```python
batting19 = pd.read_csv('batting_2019.csv')
bowling19 = pd.read_csv('bowling_2019.csv')
ipl2019 = pd.merge(batting19, bowling19, on='player')
ipl2019
```

```python
ipl2019=ipl2019[['ave','inns_x','player','sr','bowl_sr','ct','econ','inns_y','wkts']].set_index('player')
ipl2019
```

```python
ipl_19 = ipl2019.replace('-', np.nan)
ipl_19
```

```python
ipl_19['ave'] = pd.to_numeric(ipl_19['ave'])
ipl_19['inns_x'] = pd.to_numeric(ipl_19['inns_x'])
ipl_19['sr'] = pd.to_numeric(ipl_19['sr'])
ipl_19['bowl_sr'] = pd.to_numeric(ipl_19['bowl_sr'])
ipl_19['ct'] = pd.to_numeric(ipl_19['ct'])
ipl_19['econ'] = pd.to_numeric(ipl_19['econ'])
ipl_19['inns_y'] = pd.to_numeric(ipl_19['inns_y'])
ipl_19['wkts'] = pd.to_numeric(ipl_19['wkts'])
```

```python
print(ipl_19['ave'].dtype)
```

```python
ipl_19 = ipl_19.replace(np.nan,0)
ipl_19
```

```python
ipl_19['catch_val'] = ipl_19['ct']*2.5
ipl_19
```

```python
def econ_med(x):
    if x['econ'] <= ipl_19['econ'].median():
        return 2.5
    else:
        return 0
    
ipl_19['econ_val'] = ipl_19.apply(econ_med, axis=1)

ipl_19
```

```python
def sr_med(x):
    if x['sr'] >= ipl_19['sr'].median():
        return 2.5
    else:
        return 0
    
ipl_19['bat_sr_val'] = ipl_19.apply(sr_med, axis=1)

ipl_19
```

```python
ipl_19['wkts_val'] = ipl_19['wkts']*2.5
ipl_19
```

```python
def bowl_sr_med(x):
    if x['bowl_sr'] <= ipl_19['bowl_sr'].median():
        return 2.5
    else:
        return 0
    
ipl_19['bowl_sr_val'] = ipl_19.apply(sr_med, axis=1)

ipl_19
```

```python
ipl_19['value_points'] = ipl_19['ave']+ipl_19['catch_val']+ipl_19['econ_val']+ipl_19['bat_sr_val']+ipl_19['wkts_val']+ipl_19['bowl_sr_val']
```

```python
ipl_19
```

```python
ipl_19sorted = ipl_19.sort_values('value_points', ascending=False)
pd.set_option('display.max_rows', None)
ipl_19sorted
```
