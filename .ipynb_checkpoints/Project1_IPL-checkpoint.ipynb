```python
import pandas as pd
import numpy as np
```

```python
batting19 = pd.read_csv('batting_2019.csv')
batting19
```

```python
bowling19 = pd.read_csv('bowling_2019.csv')
bowling19
```

```python
ipl2019 = pd.merge(batting19, bowling19, on='player')
ipl2019
```

```python
ipl2019=ipl2019[['ave','inns_x','player','sr','bowl_sr','ct','econ','inns_y','wkts']].set_index('player')
ipl2019
```

```python
ipl_19 = ipl2019.replace('-', np.nan)
ipl_19
```

```python
ipl_19['ave'] = pd.to_numeric(ipl_19['ave'])
ipl_19['inns_x'] = pd.to_numeric(ipl_19['inns_x'])
ipl_19['sr'] = pd.to_numeric(ipl_19['sr'])
ipl_19['bowl_sr'] = pd.to_numeric(ipl_19['bowl_sr'])
ipl_19['ct'] = pd.to_numeric(ipl_19['ct'])
ipl_19['econ'] = pd.to_numeric(ipl_19['econ'])
ipl_19['inns_y'] = pd.to_numeric(ipl_19['inns_y'])
ipl_19['wkts'] = pd.to_numeric(ipl_19['wkts'])
```

```python
print(ipl_19['ave'].dtype)
```

```python
ipl_19 = ipl_19.replace(np.nan,0)
ipl_19
```

```python
ipl_19['catch_val'] = ipl_19['ct']*2.5
ipl_19
```

```python
def econ_med(x):
    if x['econ'] <= ipl_19['econ'].median():
        return 2.5
    else:
        return 0
    
ipl_19['econ_val'] = ipl_19.apply(econ_med, axis=1)

ipl_19
```

```python
def sr_med(x):
    if x['sr'] >= ipl_19['sr'].median():
        return 2.5
    else:
        return 0
    
ipl_19['bat_sr_val'] = ipl_19.apply(sr_med, axis=1)

ipl_19
```

```python
ipl_19['wkts_val'] = ipl_19['wkts']*2.5
ipl_19
```

```python
def bowl_sr_med(x):
    if x['bowl_sr'] <= ipl_19['bowl_sr'].median():
        return 2.5
    else:
        return 0
    
ipl_19['bowl_sr_val'] = ipl_19.apply(sr_med, axis=1)

ipl_19
```

```python
ipl_19['value_points'] = ipl_19['ave']+ipl_19['catch_val']+ipl_19['econ_val']+ipl_19['bat_sr_val']+ipl_19['wkts_val']+ipl_19['bowl_sr_val']
```

```python
ipl_19
```

```python
ipl_19sorted = ipl_19.sort_values('value_points', ascending=False)
pd.set_option('display.max_rows', None)
ipl_19sorted
```

```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(16,8))
ipl_19sorted.groupby('player')['value_points'].median().sort_values(ascending=False).plot.bar(color='b')
```

```python
plt.figure(figsize=(12,6))
ipl_19sorted.groupby('player')['value_points'].median().sort_values(ascending=False).head(20).plot.bar(color='b')
```

```python
import plotly
# connected=True means it will download the latest version of plotly javascript library.
plotly.offline.init_notebook_mode(connected=True)

from plotly.graph_objs import Figure, Histogram, Layout

from plotly.graph_objs import Bar

ipl_19sorted = ipl_19sorted.groupby('player')['value_points'].median().sort_values(ascending=False).head(20).reset_index()

data = [Bar(x=ipl_19sorted['player'], y=ipl_19sorted['value_points'], text=ipl_19sorted['player'])]
layout = Layout(title="IPL 2019: Players with Highest Value Points")

fig = Figure(data=data, layout=layout)

plotly.offline.iplot(fig,show_link=False)
```

```python
import pandas as pd
batting18 = pd.read_csv('ipl18bat.csv')
batting18
```

```python
bowling18 = pd.read_csv('ipl18bowl.csv')
bowling18
```

```python
ipl2018 = pd.merge(batting18, bowling18, on='player')
ipl2018
```

```python
ipl2018=ipl2018.set_index('player')
ipl2018 = ipl2018.replace('-', np.nan)

ipl2018['bat_ave'] = pd.to_numeric(ipl2018['bat_ave'])
ipl2018['bat_inns'] = pd.to_numeric(ipl2018['bat_inns'])
ipl2018['bat_mat'] = pd.to_numeric(ipl2018['bat_mat'])
ipl2018['bat_runs'] = pd.to_numeric(ipl2018['bat_runs'])
ipl2018['bat_sr'] = pd.to_numeric(ipl2018['bat_sr'])
ipl2018['bowl_inns'] = pd.to_numeric(ipl2018['bowl_inns'])
ipl2018['bowl_mat'] = pd.to_numeric(ipl2018['bowl_mat'])
ipl2018['bowl_sr'] = pd.to_numeric(ipl2018['bowl_sr'])
ipl2018['catches'] = pd.to_numeric(ipl2018['catches'])
ipl2018['econ'] = pd.to_numeric(ipl2018['econ'])
ipl2018['wkts'] = pd.to_numeric(ipl2018['wkts'])

ipl2018 = ipl2018.replace(np.nan,0)
ipl2018
```

```python
ipl2018['catch_val'] = ipl2018['catches']*2.5

def econ_med(x):
    if x['econ'] <= ipl2018['econ'].median():
        return 2.5
    else:
        return 0
    
ipl2018['econ_val'] = ipl2018.apply(econ_med, axis=1)

ipl2018['wkts_val'] = ipl2018['wkts']*2.5

def bat_sr_med(x):
    if x['bat_sr'] >= ipl2018['bat_sr'].median():
        return 2.5
    else:
        return 0
    
ipl2018['bat_sr_val'] = ipl2018.apply(bat_sr_med, axis=1)

def bowl_sr_med(x):
    if x['bowl_sr'] <= ipl2018['bowl_sr'].median():
        return 2.5
    else:
        return 0
    
ipl2018['bowl_sr_val'] = ipl2018.apply(bowl_sr_med, axis=1)
```

```python
ipl2018['value_points'] = ipl2018['bat_ave']+ipl2018['catch_val']+ipl2018['econ_val']+ipl2018['bat_sr_val']+\
ipl2018['wkts_val']+ipl2018['bowl_sr_val']

ipl2018sorted = ipl2018.sort_values('value_points', ascending=False)
```
